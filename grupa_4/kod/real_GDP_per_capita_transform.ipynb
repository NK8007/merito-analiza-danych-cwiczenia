```python
real_gdp_df_EU = real_gdp_df.drop('EA20: Euro Area (the 20 countries using the euro)', axis=1)
real_gdp_df_EU = real_gdp_df_EU.drop('EU27_2020: European Union 27 countries (after Brexit, excluding the UK)', axis=1)
```

```python
real_gdp_df_EU_rok = real_gdp_df_EU.rename({'geo\\TIME_PERIOD': 'Rok'}, axis=1)
# real_gdp_df_EU_rok
```

```python
real_gdp_df_EU_rok_short = real_gdp_df_EU_rok.rename(columns=lambda x: re.sub('[A-Z]+: ','',x))
# real_gdp_df_EU_rok_short
```

```python
real_gdp_df_EU_rok_short_noSpace = real_gdp_df_EU_rok_short.rename(columns=lambda x: re.sub(' ','_',x))
# real_gdp_df_EU_rok_short_noSpace
```
