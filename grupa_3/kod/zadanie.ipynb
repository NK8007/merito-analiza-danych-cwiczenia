```python
import pandas as pd
```

```python
path = '../../dane/real_GDP_per_capita.csv'
df_real_GDP_per_capita = pd.read_csv(path, sep=',', index_col=1)
```

```python
df_real_GDP_per_capita_1 = df_real_GDP_per_capita.copy()
df_real_GDP_per_capita_1.head()
```

Condition columns names as they contains additional and not needed elements

```python
columns_name = [x for x in df_real_GDP_per_capita_1.columns]
print(columns_name)
```

Remove addition elements like spaces and tabs from columns names

```python
df_real_GDP_per_capita_1.columns = [x.strip() for x in df_real_GDP_per_capita_1.columns]
columns_name = [x for x in df_real_GDP_per_capita_1.columns]
print(columns_name)
```

Drop unneeded columns like '#'

```python
df_real_GDP_per_capita_2 = df_real_GDP_per_capita_1.drop(columns=['#'])
df_real_GDP_per_capita_2.head()
```

Lowercase the columns name

```python
df_real_GDP_per_capita_2.columns = [x.lower() for x in df_real_GDP_per_capita_2.columns]
df_real_GDP_per_capita_2.head()
```

Transpond the df_real_GDP_per_capita_2

```python
df_real_GDP_per_capita_3 = df_real_GDP_per_capita_2.T
df_real_GDP_per_capita_3.head()
```
